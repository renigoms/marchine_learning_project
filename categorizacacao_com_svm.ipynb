### Aplicando modelo

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split as treinoteste

data_frame = data_frame.dropna() # Fazendo um drop em valores nulos
x_caracteristicas = data_frame.iloc[:, :32].values
y_previsor = data_frame.iloc[:, :1].values
```

* Gerando variÃ¡veis de treinamento

```python
x_treino, x_teste, y_treino, y_teste = treinoteste(x_caracteristicas, y_previsor, test_size=0.9)
```

* Deixando variÃ¡veis de treino e teste x equivalentes

```python
from sklearn.preprocessing import StandardScaler

sc_x_treino = StandardScaler()
sc_x_teste = StandardScaler()

x_treino = sc_x_treino.fit_transform(x_treino)
x_teste = sc_x_teste.fit_transform(x_teste)
```

* Treinando modelo

```python
svm_model = SVC(kernel='rbf')
svm_model.fit(x_treino, y_treino)
```

* Realizamdo previsÃµes

```python
previssoes = svm_model.predict(x_teste)
```

```python
previssoes
```

### Avaliando modelo

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_teste, previssoes)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_teste, previssoes))
```
