```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

```python
# load dataset
df = pd.read_csv("base/heart.csv")
print(df.columns)
```

Visualize os dados

```python
# Criando o scatter plot
plt.figure(figsize=(10, 6))  # Opcional: ajusta o tamanho do grfico
sns.scatterplot(x=df["Age"], y=df["MaxHR"], hue=df["HeartDisease"])

# Adicionando ttulos para clareza
plt.title("Relao entre Idade e Frequncia Cardaca Mxima")
plt.xlabel("Idade (Anos)")
plt.ylabel("Frequncia Cardaca Mxima (MaxHR)")

plt.show()
```

Clula para corrigir os valores de zero

```python
# Colunas com valores de zero que so clinicamente impossveis
cols_to_correct = ["RestingBP", "Cholesterol"]

# Substituir 0 por NaN (Not a Number) para que possamos calcular a mediana corretamente
for col in cols_to_correct:
    df[col] = df[col].replace(0, np.nan)

# Calcular a mediana de cada coluna (ignorando os NaN) e preencher os valores faltantes
for col in cols_to_correct:
    median_val = df[col].median()
    df[col] = df[col].fillna(median_val)

# Verificar se ainda existem valores nulos ou zeros nessas colunas
print("Valores nulos aps a correo:")
print(df[cols_to_correct].isnull().sum())
print("\nNovas estatsticas descritivas:")
print(df[cols_to_correct].describe())
```

Etapa 1: Aplicar o One-Hot Encoding

```python
# Primeiro, identificamos as colunas que so realmente categricas
# Note que no inclumos a varivel alvo 'HeartDisease'
categorical_features = [
    "Sex",
    "ChestPainType",
    "FastingBS",
    "RestingECG",
    "ExerciseAngina",
    "ST_Slope",
]

# Usamos pd.get_dummies para criar as novas colunas
# drop_first=True: Remove a primeira categoria de cada feature. Isso evita redundncia (multicolinearidade).
# dtype=int: Cria as novas colunas com 0s e 1s em vez de True/False.
df_encoded = pd.get_dummies(
    df, columns=categorical_features, drop_first=True, dtype=int
)

# Vamos ver como ficou o novo DataFrame com as colunas transformadas
print("Colunas do DataFrame aps o One-Hot Encoding:")
print(df_encoded.columns)
print("\nPrimeiras linhas do novo DataFrame:")
print(df_encoded.head())
```

Etapa 2: Nova Seleo de Recursos e Diviso
Seleo de recurso

- Variveis independentes (features):
So as colunas que voc usa como base para prever a target.

- Varivel dependente (target):
A varivel que voc quer prever  neste caso, se a pessoa tem ou no doena cardaca

```python
# Agora, separamos X e y a partir deste NOVO DataFrame codificado
X = df_encoded.drop("HeartDisease", axis=1)  # remove a coluna target, fica s com as features
y = df_encoded["HeartDisease"]  # pega s a coluna target para prever

# O restante do processo continua igual!
# Dividimos os novos X e y em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=16
)

# Aplicamos a padronizao
scaler = StandardScaler()

X_train_scaled = pd.DataFrame(
    scaler.fit_transform(X_train), columns=X_train.columns, index=X_train.index
)
X_test_scaled = pd.DataFrame(
    scaler.transform(X_test), columns=X_test.columns, index=X_test.index
)
```

Otimizao do Hiperparmetro K para o KNN

```python
# Encontrando o melhor valor de 'k'

k_values = range(1, 31)
accuracy_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    y_pred_knn = knn.predict(X_test_scaled)
    accuracy_scores.append(accuracy_score(y_test, y_pred_knn))
```

Plotando os resultados para encontrar o melhor valor de k

```python
plt.figure(figsize=(12, 6))
plt.plot(
    k_values,
    accuracy_scores,
    color="blue",
    linestyle="dashed",
    marker="o",
    markerfacecolor="red",
    markersize=10,
)
plt.title("Acurcia vs. Valor de K")
plt.xlabel("K")
plt.ylabel("Acurcia")
plt.grid(visible=True)
plt.show()
```

Avaliando o modelo com o melhor 'k'
Pelo seu grfico, K=12  uma excelente escolha

```python
best_index = np.argmax(accuracy_scores)
best_k = k_values[best_index]
print(f"Best 'K': {best_k}")
```

```python
best_k = 13 # Atinge um recall maior que 12

knn_final = KNeighborsClassifier(n_neighbors=best_k)
knn_final.fit(X_train_scaled, y_train)
y_pred_knn_final = knn_final.predict(X_test_scaled)

print(f"\nAvaliao do Modelo KNN com k={best_k}:\n")
print(classification_report(y_test, y_pred_knn_final, target_names=["Sem Doena", "Com Doena"]))
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred_knn_final)
cnf_matrix
```
